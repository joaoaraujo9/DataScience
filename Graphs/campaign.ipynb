```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
from plotly.offline import init_notebook_mode
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output

data = pd.read_csv('C:/Users/joao_/Desktop/4 ANO JOAO/SINO/VS CODE PROJETO/Datasets/bank-full-transformed.csv')
```

```python
lst = [data]

for column in lst:
    column.loc[column["campaign"] == 1,  'campaign'] = "Uma vez"
    column.loc[column["campaign"] == 2,  'campaign'] = "Mais do que uma vez"
```

```python
px.histogram(data, x="campaign", color="campaign") 
```

```python
px.histogram(data, x="campaign", color="y", barmode = 'group') 
```

```python
campaigny = pd.DataFrame(data.groupby('campaign')['y'].sum()).reset_index()
campaigny = campaigny.sort_values(by = 'y', ascending = False)
fig = px.bar(campaigny.iloc[:], #plotting only the top 20 Countries
             x = 'campaign', y = 'y', title = 'Distribuio de Subscrio com Nmero de vezes que foram contactados')

fig.show()
```

```python
count_job_response_pct = pd.crosstab(data['y'],data['campaign']).apply(lambda x: x/x.sum() * 100)
count_job_response_pct = count_job_response_pct.transpose()
```

```python
plot_job = count_job_response_pct[1].sort_values(ascending = True).plot(kind ='barh',figsize = (12,6))
                                                                               
plt.title('Taxa de Subscrio por nmero de contactos')
plt.xlabel('Taxa de Subscrio')
plt.ylabel('Nmero de contactos')

# Label each bar
for rec, label in zip(plot_job.patches,
                      count_job_response_pct[1].sort_values(ascending = True).round(1).astype(str)):
    plot_job.text(rec.get_width()+0.8, 
                  rec.get_y()+ rec.get_height()-0.5, 
                  label+'%', 
                  ha = 'center', 
                  va='bottom')
```

```python
labels = data['campaign'].value_counts().index
values = data['campaign'].value_counts().values

fig = px.pie(data, values= values, names= labels)
fig.update_traces(textposition='inside')
fig.update_layout(uniformtext_minsize=12, uniformtext_mode='show')
fig.show()
```
